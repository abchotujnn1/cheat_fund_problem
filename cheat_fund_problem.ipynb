```python
def calculate(data,month):
    Month=np.array(data["Month"])
    contribution=np.array(data["Contribution"])
    bid_w=np.array(data["Amount won by the bidder"])
    fun_com=np.array(data["Chit fund organizer commission"])
    net_bit_w=np.array(data["Net amount recd by Bid winner"])
    amt_to_e=np.array(data["Amount returned to everyone in the group"])
    if(month==24):
    #Annualized  Return of the person who bids in the last month 25
       amt=bid_w[month]-fun_com[0]+sum(amt_to_e)
    
    # Annualized Return of the person who bids in the first 
    if(month==0):
        amt=bid_w[month]-fun_com[0]+sum(amt_to_e)
     #annualized return of chit fund participant 
    data["annualized_return"]=[bid_w[i]-fun_com[i]+sum(amt_to_e) for i in (Month-1)]
    data["Return % for each month's bid winner"]=[(((bid_w[i]-fun_com[i]+sum(amt_to_e))-sum(contribution))/sum(contribution))*100 for i in (Month-1)]
    return amt, data
    
```

```python
amt,data=calculate(data,0)
print(amt)
```
